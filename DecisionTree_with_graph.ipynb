```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from matplotlib import pyplot
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
col_names = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
pima = pd.read_csv('/content/Iris.csv')
pima.head(10)
```

```python
feature_cols = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
X = pima[feature_cols]
y = pima.Species
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

Calling the DecisionTreeClassifier to make a decision tree for the dataset

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:" , metrics.accuracy_score(y_test, y_pred))
```

```python
X, y = make_classification(n_samples=10000, n_features=20, n_informative=5, n_redundant=15, random_state=5)

print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
train_scores, test_scores = list(), list()

values = [i for i in range(1,21)]
```

```python
for i in values:
  model = DecisionTreeClassifier(max_depth=i)
  model.fit(X_train, y_train)
  train_yhat = model.predict(X_train)
  train_acc = accuracy_score(y_train, train_yhat)
  train_scores.append(train_acc)

  test_yhat = model.predict(X_test)
  test_acc = accuracy_score(y_test, test_yhat)
  test_scores.append(test_acc)

  print('>%d, train: %.3f, test: %.3f' % (i, train_acc, test_acc))
```

```python
pyplot.plot(values, train_scores, '-o', label='Train')
pyplot.plot(values, test_scores, '-o', label='Test')
pyplot.legend()
pyplot.show()
```

From the above graph we see that around the 7.5 mark the training and testing scores dtar to deviate showing the point of overfitting

```python
data = pd.read_csv('/content/carprediction (1).csv')
data.head(10)
```

```python
data.info()
```

```python
null_rows = data.isna().sum()
```

```python
X = data.drop('MSRP', axis=1)
y = data['MSRP']
```

```python
categorical_cols = X.select_dtypes(include='object').columns
```

```python
for col in categorical_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

X = X.fillna(X.mean())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
model = DecisionTreeRegressor(random_state = 10)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
```

```python
max_depth_values = range(1, 21)

train_errors = []
test_errors = []

for max_depth in max_depth_values:
    dt_regressor = DecisionTreeRegressor(max_depth=max_depth, random_state=42)
    dt_regressor.fit(X_train, y_train)

    y_train_pred = dt_regressor.predict(X_train)
    y_test_pred = dt_regressor.predict(X_test)

    train_error = mean_squared_error(y_train, y_train_pred)
    test_error = mean_squared_error(y_test, y_test_pred)

    train_errors.append(train_error)
    test_errors.append(test_error)
```

```python
for i in range(len(max_depth_values)):
    train_error = train_errors[i]
    test_error = test_errors[i]

    print(f">Train: {train_error:.3f} >Test: {test_error:.3f}")
```

```python
plt.figure(figsize=(5, 6))
plt.plot(max_depth_values, train_errors, label='Training Error', marker='o')
plt.plot(max_depth_values, test_errors, label='Testing Error', marker='o')
plt.xlabel('Max Depth of Decision Tree')
plt.ylabel('Mean Squared Error')
plt.title('Training and Testing Errors vs. Model Complexity')
plt.legend()
plt.show()
```

From the above graph we see that around the 5 mark the training and testing scores dtar to deviate showing the point of overfitting
